```python
while True:
    try:
        idade = int(input("Por favor, digite a idade do cliente: "))

        if idade < 16:
            print("ENTRADA PERMITIDA APENAS COM ACOMPANHANTE")
        else:
            valor_carteira = float(input("Por favor, digite o valor que o cliente tem na carteira: "))
            if valor_carteira >= 30:
                print("ENTRADA LIBERADA! BOM FILME")
            else:
                print("SALDO INSUFICIENTE PARA O INGRESSO")

    except ValueError:
        print("Saldo insuficiente para o ingresso")

    continuar = input("Deseja verificar outro cliente? (S/N): ").strip().upper()
    if continuar == 'N':
        break
    elif continuar == 'S':
        print("Programa encerrado.")
```
